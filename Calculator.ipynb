#Basic Calculator Using *Python*

```python
while True:
  a = int(input("Enter your first number:"))
  b = int(input("Enter your second number:"))

  sign = input("What Operation would you like to perform?(+ , - , * , / , ** ,  %):")
  print(type(sign))

  if(sign == '+'):
    print('Addition of the numbers:', a+b)
  elif(sign == '-'):
    print('Substraction of the numbers:', a-b)
  elif(sign == '*'):
    print('Multiplucation of the numbers:', a*b)
  elif(sign == '/'):
    print('Division of the numbers:', a/b)
  elif(sign == '**'):
    print('Power of a:', a**b)
  elif(sign == '%'):
    print('Remainder of the number (a):', a%b)


```
